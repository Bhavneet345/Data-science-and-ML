### Importing the libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
```

### Importing the dataset

```python
data = pd.read_csv('Market_Basket_Optimisation.csv',header=None)
data.head()
```

### Data Preprocessing

```python
col = []
for i in range (1,np.shape(data)[1] + 1):
    col.append('Product'+str(i))
data.columns = col
data.head()
```

Changing the format of dataframe as per ''apyori'' module

```python
trans =[]
for i in range(7501):
    trans.append([str(data.values[i,j]) for j in range(0,20)])
```

### Training the Apriori model on training set

```python
from apyori import apriori
# Arguments dependes upon the business problem
rules = apriori(transactions = trans, min_support = 0.003, min_confidence = 0.2, min_lift = 3, min_length = 2,
               max_length = 2)
```

### Visulalising the results

```python
results = list(rules)
```

```python
for items in results:
    print(items)
```

### Putting the rules in Pandas DataFrame

```python
def inspect(result):
    Left_side = [tuple(product[2][0][0])[0] for product in result]
    Right_side = [tuple(product[2][0][1])[0] for product in result]
    supports = [product[1] for product in result]
    confidence = [product[2][0][2] for product in result]
    lift = [product[2][0][3] for product in result]
    return zip(Left_side,Right_side,supports,confidence,lift)
rules_dataframe = pd.DataFrame(inspect(results),columns = ['Left_side', 'Right_side', 'support', 'confidence', 'lift'])
```

```python
rules_dataframe
```

### Sorting the results

```python
rules_dataframe.sort_values(['lift'],ascending=False,inplace=True)
rules_dataframe.reset_index(drop=True,inplace=True)
```
