### Importing the libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### Importing the dataset

```python
data = pd.read_csv("Position_Salaries.csv")
X = data.iloc[:,1:-1].values
y = data.iloc[:,-1:].values
```

### Training the Linear Regression model

```python
from sklearn.linear_model import LinearRegression
lin_regressor = LinearRegression()
lin_regressor.fit(X,y)
```

```python
lin_regressor.predict(X)
```

### Training the Polynomial Regression model

```python
### Create the polynomial feature with degree 2 or greater and then fit into linear regression..
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree=2)
X_poly = poly_reg.fit_transform(X)
### Fitting into the linear regression model
poly_regressor = LinearRegression()
poly_regressor.fit(X_poly,y)
```

```python
poly_regressor.predict(X_poly)
```

### Visualising the Linear regression model

```python
pred1 = lin_regressor.predict(X)
plt.scatter(X,y,color="blue",label="Orignal")
plt.plot(X,pred1,"g",label="predicted")
plt.xlabel("level")
plt.ylabel("salary")
plt.legend()
plt.show()
```

### Visuallising the polynomial model

```python
pred2 = poly_regressor.predict(X_poly)
plt.scatter(X,y,color="blue",label="orignal")
plt.plot(X,pred2,"g",label="predicted")
plt.xlabel("level")
plt.ylabel("salary")
plt.legend()
plt.show()
```

### Visualising the Polynomial regression results (for higher resolutions)

```python
### From class PolynomialFeatures
poly2 = PolynomialFeatures(degree=4)
X_poly2 = poly2.fit_transform(X)
poly_reg2 = LinearRegression()
poly_reg2.fit(X_poly2,y)
pred3 = poly_reg2.predict(X_poly2)
plt.scatter(X,y,label="Orignal",color="blue")
plt.plot(X,pred3,"g",label="Predicted")
plt.xlabel("level")
plt.ylabel("Salary")
n = np.arange(1,10,0.5)
plt.xticks(n)
plt.legend()
plt.show()
###PERFECT FIT
```

### Predicting the new results with linear regression

```python
new=np.array([[6.5]])
new_pred = lin_regressor.predict(new)
print(f"The salary of level 6.8 should be around {(new_pred)}")
```

### Predicting the new results with Polynomial regression

```python
new1 = np.array([[6.5]])
new1 = poly2.fit_transform(new1)
x = poly_reg2.predict(new1)
print(f"The salary of level 6.8 should be around {((x))}")
```
