### Importing the libraries

```python
import pandas as pd
import numpy as np
```

### Import datasets

```python
data = pd.read_csv("50_Startups.csv")
y = data.iloc[:,:-1]
w = data.iloc[:,-1:]
```

### Encoding the categorical data

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
ct = ColumnTransformer(transformers=[("encoder",OneHotEncoder(),[3])],remainder="passthrough")
y = np.array(ct.fit_transform(y)).astype("int32")
```

### Adding the one's column for backward elimination

```python
# y = np.append(arr=np.ones((50,1)).astype("int32"),values=y,axis=1)
```

### Backward elimination..

```python
import statsmodels.api as sm 
X_opt = y[:,0:6]
regressor_OLS = sm.OLS(w,X_opt).fit()
regressor_OLS.summary()
```

### Removing the variable with highest p-value (X1) and fitting the variables again

```python
import statsmodels.api as sm 
X_opt = y[:,[0,2,3,4,5]]
regressor_OLS = sm.OLS(w,X_opt).fit()
regressor_OLS.summary()
```

```python
import statsmodels.api as sm 
X_opt = y[:,[0,2,3,4]]
regressor_OLS = sm.OLS(w,X_opt).fit()
regressor_OLS.summary()
```

```python
import statsmodels.api as sm 
X_opt = y[:,[0,2,3]]
regressor_OLS = sm.OLS(w,X_opt).fit()
regressor_OLS.summary()
```
